### Naver Hotel data crawling 

```python
import pandas as pd
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
import time

service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

hotel_name = []
hotel_naver_url=[]

for page in range(23):  
    url = f"https://hotels.naver.com/list?placeFileName=place%3A
    Seoul&adultCnt=2&includeTax=false&propertyTypes=0&pageIndex={page}"
    driver.get(url)
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    content_all=soup.find(class_="SearchList_SearchList__PL2mv")
    lis = content_all.find_all("li")

    for content in lis:
        name = content.find("h4",class_="Detail_title__9rvke").text
        hotel_name.append(name)
        
        link = content.find("a",
                            class_="SearchList_anchor__zbpU4").attrs["href"].split("?")[1]
        final_link = "https://hotels.naver.com/item?"+link
        hotel_naver_url.append(final_link)

driver.quit()
```

```python
print(len(hotel_name))
print(len(hotel_naver_url))

final = pd.DataFrame({"상호명":hotel_name,"콘텐츠URL":hotel_naver_url,"서비스유행":"hotel"})
final.to_excel("./서울호텔.xlsx",index=False)    
final
```
